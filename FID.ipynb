```python
fid_resluts =[]
```

```python
#paths
new_dit_img_path = "/content/drive/MyDrive/colab_new_dit/samples_20251205_"
dit_img_path = "/content/drive/MyDrive/colab_dit/samples_org_20251205_"
imgnet_path = "/content/drive/MyDrive/imagenet/data/00"
```

```python
from cleanfid import fid
```

```python
for cls in [972,973,974,975,976]:
  cls = str(cls)
  score_new_dit = fid.compute_fid(new_dit_img_path + cls, imgnet_path + cls)
  score_dit = fid.compute_fid(dit_img_path + cls, imgnet_path + cls)
  fid_resluts.append({cls:{"new_dit":score_new_dit, "org_dit":score_dit}})
  print(f"score_new_dit: {score_new_dit}, org dit: {score_dit}")
```

```python
kid_results = []
```

```python
for cls in [972,973,974,975,976]:
  cls = str(cls)
  score_new_dit = fid.compute_kid(new_dit_img_path + cls, imgnet_path + cls)
  score_dit = fid.compute_kid(dit_img_path + cls, imgnet_path + cls)
  kid_results.append({cls:{"new_dit":score_new_dit, "org_dit":score_dit}})
  print(f"\n\n score_new_dit: {score_new_dit}, org dit: {score_dit}\n")
```

```python
# save results to json file
import json
with open('/content/drive/MyDrive/colab_dit_results/kid_results_cls_972_to_976_2025_12_05.json', 'w') as fp:
    json.dump(kid_results, fp)
```
